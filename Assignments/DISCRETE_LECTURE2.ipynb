Consider the following experiment:
1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$ is $p_i$, where $i = 1, 2, ..., k$, and remains the same from trial to trial. Note that $\sum_{i=1}^k p_i = 1$.
4. The trials are independent.
5. The random variables of interest are $Y_1, Y_2, ..., Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Note that $\sum_{i=1}^k Y_i = n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdot P(y) = n \cdot p_i$

Step 1:
The expected value of $Y_i$ is given by $E[Y_i] = \sum_{y} y \cdot P(y)$

Step 2:
Using the joint probability distribution:
$E[Y_i] = \sum_{y} y \cdot P(y) = \sum_{y} y \cdot \frac{n!}{y_1!y_2!y_3!\cdots y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}\cdots P_n^{y_n}$

Step 3:
Now, $Y_i$ is the number of trials where the outcome of each trial falls into cell $i$.

Step 4:
$Y_i$ will be equal to $n \cdot p_i$, where $p_i$ is the probability of the outcome falling into cell $i$ in a trial.

Step 5:
$E[Y_i] = n \cdot p_i$. **QED**

Step 3:
The sum of all probabilities in a probability distribution is always equal to 1. Using this, rewrite:

$$ E(Y) = \sum_{y=1}^{\infty} y \cdot p(y) + 0 \cdot \frac{\lambda^0}{0!} e^{-\lambda} $$

Step 4:
Simplify the terms:

$$ E(Y) = e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} $$
