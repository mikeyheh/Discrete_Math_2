Simple Graph
==
A simple graph is an undirected graph that has no loops or multiple edges. In the diamond shaped graph shown below is an example of a simple graph, as there are no loops nor parallel edges.

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4])
edgelist = [(1,2),(2,3),(3,4),(4,1)]
G.add_edges_from(edgelist)
pos = nx.circular_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

Bipartite Graph
==
A Bipartite Graph is a graph where the vertices are in two different sets such that all vertex are connected to another vertex from a different set.

```python
import networkx as nx
import matplotlib.pyplot as plt

G= nx.Graph()
G.add_nodes_from([1, 2, 3, 4], bipartite=0)
G.add_nodes_from(['a', 'b', 'c'], bipartite=1) 

G.add_edges_from([(1, 'a'), (1, 'b'), (2, 'b') , (3, 'c'),(3, 'b'), (4, 'c')])

color = ['red' if G.nodes[n]['bipartite'] == 0 else 'black' for n in G.nodes]

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G,top)
nx.draw(G, pos=pos, with_labels = True, node_color=color, font_color='black')
plt.show()
```

Subgraphs

A subraph is a graph whose vetrices and edges are contained in a larger graph

Travelling in a Graph
=

Walk
-
A walk in graph theory refers to the traversing of a graph. In this method of traversing edges and vertices <b>CAN</b> be repeated

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,3),(4,5),(5,1)]
G.add_edges_from(edgelist)
nx.draw(G,with_labels=True)
plt.show()
```

Route: $1 \to  3 \to 4 \to 5 \to 1 \to 2$

Trail
-
Just like a walk, a trail is a way of traversing or travelling the graph, however unlike a walk, trails must have no repeated edge.

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,3),(4,5),(5,1)]
G.add_edges_from(edgelist)
nx.draw(G,with_labels=True)
plt.show()
```

Route: $2 \to  1 \to 3 \to 4 \to 1 \to 5$

Path
-
A path is a walk in which no vertex is repeated. In other words, you cannot traverse a vertex more than once.

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,5),(5,1)]
G.add_edges_from(edgelist)
nx.draw(G,with_labels=True)
plt.show()
```

Route: $2 \to  3 \to 1 \to 5 \to 4$

Closed Walk
-
A closed walk has the same concepts as a regular walk, however the route must end with the same vertex it started.

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,5),(5,1)]
G.add_edges_from(edgelist)
pos=nx.circular_layout(G)
nx.draw(G,pos=pos,with_labels=True)
plt.show()
```

Route: $5 \to  4 \to 1 \to 3 \to 2 \to 1 \to 5$

Euler's Circuit
- 
Euler's circuit is a circuit in which every edge must be visted at exactly once. Conditions must be met in order to have a Eulerian circuit, one of the conditions is that every vertex in the graph has an even degree, and the other is that the graph must be connected.


```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,5),(5,1)]
G.add_edges_from(edgelist)
pos=nx.planar_layout(G)
nx.draw(G,pos=pos,with_labels=True)
plt.show()
```

Travelling Salesman Problem
-
It is an alogrithmic problem tasked with finding the shortest route between a set of points and locations that must be visted.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E','F']
G.add_nodes_from(nodes)

# Add edges with weights (distances)
edgelist = [
    ('A', 'B', 10),
    ('A', 'C', 15),
    ('A', 'E', 35),
    ('B', 'C', 10),
    ('B', 'D', 20),
    ('B', 'E', 5),
    ('C', 'E', 25),
    ('D', 'E', 30),
    ('D', 'C', 10),
    ('F', 'A', 25),
    ('F', 'B', 15),
    ('F', 'C', 20),
    ('F', 'D', 5),
    ('F', 'E', 30)
]

G.add_weighted_edges_from(edgelist)

# Set up the plot
plt.figure(figsize=(8, 6))

# Use a circular layout for the graph
pos = nx.circular_layout(G)

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)

# Draw the edges
nx.draw_networkx_edges(G, pos, edgelist=G.edges())

# Draw the labels on the nodes
nx.draw_networkx_labels(G, pos)


edge_labels = {(u, v): f'{d["weight"]} km' for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)


plt.title("Graph for Travelling Salesman Problem", fontsize=15)
plt.axis('off')  
plt.show()
```
